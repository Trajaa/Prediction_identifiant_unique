```python
import pandas as pd

# Colonnes pertinentes pour le dédoublonnage
colonnes_pertinentes = ['profile_id', 'subprofile', 'produit', 'full_vol',
       'full_secteur', 'full_city']

# Critères de dédoublonnage
critères_dédoublonnage = data[colonnes_pertinentes]

# Identifier les doublons
doublons = data.duplicated(subset=critères_dédoublonnage, keep=False)

# Filtrer les doublons identifiés
doublons_df = data[doublons]

# Regrouper les doublons par ensemble
ensembles_doublons = doublons_df.groupby(colonnes_pertinentes)

# Afficher chaque ensemble de doublons
for ensemble, groupe in ensembles_doublons:
    print("Ensemble de doublons :")
    print(groupe)
    print("----------------------------------------")
```
