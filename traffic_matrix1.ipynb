```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
from scipy.spatial.distance import cdist

# Slectionner les colonnes pertinentes pour la mthode Elbow
# Ajoutez les noms des colonnes supplmentaires  la liste features
features = data[['a_own', 'b_number', 'duration', 'duration_wh',
       'num_interactions', 'num_interactions_wh', 'num_interactions_sms',
       'num_interactions_sms_wh', 'offnet']]

# Normaliser les donnes si ncessaire (utilisez StandardScaler ou MinMaxScaler)

# Appliquer la mthode Elbow pour dterminer le nombre optimal de clusters
distortions = []
K = range(1, 10)  # Essayer diffrents nombres de clusters
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(features)
    kmeanModel.fit(features)
    distortions.append(sum(np.min(cdist(features, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / features.shape[0])

# Tracer la courbe de la variance explique en fonction du nombre de clusters
plt.plot(K, distortions, 'bx-')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Distorsion')
plt.title('Mthode Elbow')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from scipy.spatial.distance import cdist

# Slectionner les colonnes pertinentes pour la mthode Elbow
# Ajoutez les noms des colonnes supplmentaires  la liste features
selected_columns = ['a_own', 'b_number', 'duration', 'duration_wh',
                    'num_interactions', 'num_interactions_wh', 'num_interactions_sms',
                    'num_interactions_sms_wh', 'offnet']
features = data[selected_columns]

# Convertir les colonnes au bon type de donnes (numrique)
features = features.astype(float)

# Appliquer la mthode Elbow pour dterminer le nombre optimal de clusters
distortions = []
K = range(1, 10)  # Essayer diffrents nombres de clusters
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(features)
    kmeanModel.fit(features)
    distortions.append(sum(np.min(cdist(features, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / features.shape[0])

# Tracer la courbe de la variance explique en fonction du nombre de clusters
plt.plot(K, distortions, 'bx-')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Distorsion')
plt.title('Mthode Elbow')
plt.show()
```

```python
from sklearn.cluster import KMeans
import pandas as pd

# Slection des colonnes pertinentes
selected_columns = ['a_own', 'b_number', 'duration', 'duration_wh',
                    'num_interactions', 'num_interactions_wh', 'num_interactions_sms',
                    'num_interactions_sms_wh', 'offnet']

# Extraction des donnes  partir des colonnes slectionnes
X = data[selected_columns]

# Instanciation et ajustement du modle K-means
k = 3  # Nombre de clusters souhait
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)

# Rcupration des tiquettes de cluster
labels = kmeans.labels_

# Rcupration des coordonnes des centres de cluster
cluster_centers = kmeans.cluster_centers_

# Affichage des tiquettes de cluster pour chaque point de donnes
print(labels)

# Affichage des coordonnes des centres de cluster
print(cluster_centers)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score

# Calcul du score de silhouette pour l'ensemble des donnes
silhouette_avg = silhouette_score(X, labels)
print("Le score de silhouette moyen est :", silhouette_avg)
```
