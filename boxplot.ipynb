```python
import pandas as pd

# Chargement du jeu de données à partir d'un fichier CSV (par exemple)
data = pd.read_csv('C:/Users/hp/Desktop/prj_pfe_inwi/top_hier_bts_main_d1_10000_chiffre.csv')
 
```

```python
columns_to_drop1 =['full_region','id_date','imei','segment']
data1= data.drop(columns=columns_to_drop1)
data1 = data1.drop(data.columns[0], axis=1) 
print(data1.columns)
```

```python
import pandas as pd

# Filtrer les données avec source = 'voice'
voice_data = data1.loc[data1['source'] == 'VOICE']

# Filtrer les données avec source = 'data'
data_data = data1.loc[data1['source'] == 'DATA']

# Afficher les données avec source = 'voice'
print("Voice Data:")
print(voice_data)

# Afficher les données avec source = 'data'
print("Data Data:")
print(data_data)
```

```python
import matplotlib.pyplot as plt

# Créer une figure et des axes
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
axes[0].boxplot(data_data["full_vol"])
axes[0].set_title("Boîte à moustaches - data_data full_vol")
# Tracer la boîte à moustaches pour la colonne "full_vol"
axes[1].boxplot(voice_data["full_vol"])
axes[1].set_title("Boîte à moustaches - voice_data full_vol")

# Afficher la figure
plt.show()
```
